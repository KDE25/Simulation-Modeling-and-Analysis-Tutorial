# Hyperband

Hyperband extends Successive Halving:

* Uses multiple brackets
* Allocates resources differently in each bracket
* Better exploration of large search spaces
* Commonly used in deep learning

Hyperband automatically balances exploration and exploitation.

# Load Dataset

The dataset is separated into:

X → features

y → classification target

The goal is to provide data for the model training and evaluation process.

```python
data = load_breast_cancer()
X = data.data
y = data.target
```

# Define Model Builder

```python
def build_model(hp):
    model = keras.Sequential()

    model.add(keras.layers.Dense(
        units=hp.Int("units", min_value=16, max_value=128, step=16),
        activation="relu"
    ))

    model.add(keras.layers.Dense(1, activation="sigmoid"))

    model.compile(
        optimizer=keras.optimizers.Adam(
            hp.Choice("learning_rate", [1e-2, 1e-3, 1e-4])
        ),
        loss="binary_crossentropy",
        metrics=["accuracy"]
    )

    return model
```

# Define Hyperband Tuner

```python
tuner = kt.Hyperband(
    build_model,
    objective="val_accuracy",
    max_epochs=30,
    factor=3,
    directory="hyperband_dir",
    project_name="breast_cancer_lab"
)
```

# Run Hyperband Search

```python
tuner.search(
    X, y,
    epochs=30,
    validation_split=0.2,
    verbose=1
)

best_hp = tuner.get_best_hyperparameters(1)[0]

print("Best Units:", best_hp.get("units"))
print("Best Learning Rate:", best_hp.get("learning_rate"))
```
